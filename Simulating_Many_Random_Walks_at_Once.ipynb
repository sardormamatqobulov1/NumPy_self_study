```python
import numpy as np
```

```python
nwalks = 5000
```

```python
nsteps = 1000
```

```python
rng = np.random.default_rng()
```

```python
draws = rng.integers(0,2, size=(nwalks, nsteps))
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walks = steps.cumsum(axis=1)
```

```python
walks
```

```python
walks.max()
```

```python
walks.min()
```

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)
```

```python
hits30
```

```python
hits30.sum()
```
