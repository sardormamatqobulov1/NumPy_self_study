```python
import numpy as np
```

```python
arr = np.arange(10)
np.save("some_array", arr)
```

```python
np.load("some_array.npy")
```

```python
np.savez("array_archive.npz", a=arr, b=arr)
```

```python
arch=np.load("array_archive.npz")
```

```python
arch["b"]
```

```python
np.savez_compressed("arrays_compressed.npz")
```
