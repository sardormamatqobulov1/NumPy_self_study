##Boolen Indexing

```python
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
```

```python
data = np.array([[4,7], [0,2], [-5, 6], [0,0], [1,2], [-12, -4], [3,4]])
```

```python
names
```

```python
data
```

```python
names == "Bob"
```

```python
data[names == "Bob"]
```

```python
data[names == "Bob", 1:]
```

```python
data[names == "Bob", 1:]
```

```python
data[names == "Bob", 1]
```

```python
names != "Bob"
```

```python
~(names == "Bob")
```

```python
data[~(names == "Bob")]
```

```python
cond = names == "Bob"
```

```python
data[~cond]
```

```python
mask = (names == "Bob") | (names == "Will")
```

```python
mask
```

```python
data[mask]
```

```python
data[data < 0] = 0
```

```python
data
```

```python
data[names != "Joe"] = 7
```

```python
data
```

##Fancy Indexing

```python
arr
```

```python
arr[[4,2,0,6]]
```

```python
arr[[-3, -5, -7]]
```
